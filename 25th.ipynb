```python
def avg_new(*args):
    total = 0
    for num in args:
        total += num
    return total / len(args)
    
```

```python
avg_new(2,4)
```

```python
my_num = [2,3,4,5,6]
avg_new(*my_num)
```

```python
def f(x, y, *args, **kwargs):
    print(x, y)
    print(f'args:   {args}')
    print(f'kwargs: {kwargs}')
```

```python
f(2,3,4,5,6, a=6, b=99, c=999)
```

```python
def operation(x,y, op='+'):
    if op=='+':
        return x + y
    elif op=='-':
        return x - y
    elif op=='*':
        return x * y
```

```python
operation(3,2, op='*')
```
