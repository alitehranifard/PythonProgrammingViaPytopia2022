```python
class Human:
    def __init__(self, firstname, lastname):
        self.firstname = firstname
        self.lastname = lastname
    
    def speak(self, text):
        return f"{self.firstname} says: {text}"
```

```python
class Parent(Human):
    def __init__(self, firstname, lastname):
        self.firstname = firstname
        self.lastname = lastname
        self.kids = []
    
    def have_child(self, child_name):
        print(f'{self.firstname} is having a child name {child_name}')
        child = Child(self, child_name)
        self.kids.append(child)

        return child
    
    def __str__(self):
        return f'Parent: {self.firstname} {self.lastname}'
```

```python
class Child(Human):
    def __init__(self, parent, firstname):
        self.parent = parent
        self.firstname = firstname
        self.lastname = parent.lastname
    def __str__(self):
        return f'Child: {self.firstname} {self.lastname} of {self.parent!s}'
```

```python
p = Parent('Meysam', 'Shiri')
```

```python
p.speak('Hi there')
```

```python
hosein = p.have_child('Hosein')
```

```python
print(hosein)
```

```python
print(p)
```

```python
hosein.parent.firstname
```

Overriding

```python
from itertools import zip_longest
class Point:
    def __init__(self, *args):
        self.args = args
    
    def __abs__(self):
        sum_values = sum([item ** 2 for item in self.args])
        return sum_values ** (1 / len(self.args))
    
    def __gt__(self, obj):
        return abs(self) > abs(obj)

    def __repr__(self):
        return f"{self.__class__.__name__}{self.args}"
    
    def __add__(self, other):
        args = map(sum, zip_longest(self.args, other.args, fillvalue=0))
        return Point(*args)
```

```python
p = Point(3, 4)
```

```python
abs(p)
```

```python
p1 = Point(3, 4)
p2 = Point(6, 8)
```

```python
p1 > p2
```

```python
p1 + p2 
```

```python
p3 = Point(3, 5, 7)
p4 = Point(5, 7, 9)
```

```python
p3 + p4
```

```python
import string
string.punctuation
punc_list = string.punctuation
```

```python
def preprocess_text(text):
    text = text.lower()
    text = ''.join(filter(lambda char: not char.isdigit(), text))
    text = ''.join(filter(lambda char: not char in punc_list, text) )
    text = ' '.join(text.split())
    return text
```
