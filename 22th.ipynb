```python
artist_dict = {
    'Shadmehr' : '3k',
    'Andi' : '15k',
    'Erfan': '12k',
}
```

```python
while artist != 'quit':
    if artist in artist_dict:
        print(f'the number of sales for {artist}: {artist_dict[artist]}')
    else :
        print('No sales record in database!')
        records = input(f'please enter number of sales for {artist}')
        artist_dict[artist] = records
        print('Record updated successfully')
    artist = input('Enter another artist name')
```
