```python
class PeopleOfIran:
    # class attributes
    nationality = "Iranian"
    native_language = "Persian"
    ## dunder method
    def __init__(self, name, age, city):
        # instance attributes
        self.name = name
        self.age = age
        self.city = city
    def speak(self):
        # method
        return f"{self.name} says: Salam!"
    def introduce(self):
        # method
        return f"My name is {self.name}, I am {self.age} years old and I live in {self.city}."
```

```python
ali = PeopleOfIran("Ali", 25, "Tehran")
# instance of PeopleOfIran named ali
```
