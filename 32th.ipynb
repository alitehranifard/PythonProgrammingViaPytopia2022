```python
from abc import ABC, abstractmethod
```

```python
class Shape(ABC):
    def __init__(self, color='Black'):
        self.color = color
    
    @abstractmethod
    def area(self):
        pass
    @abstractmethod
    def perimeter(self):
        pass
```

```python
class Rectangle(Shape):
    def area(self):
        return 10
    def perimeter(self):
        return 20
```

```python
class Point:
    def __init__(self, *args):
        self.vector = args
    
    def __str__(self):
        return f"{self.__class__.__name__}({self.vector})"
    def __repr__(self):
        return f'{self.__class__.__name__}({self.vector})'
    def __abs__(self):
        sum_value = sum([e ** 2 for e in self.vector])
        return sum_value ** 0.5
    def __lt__(self, obj):
        return abs(self) < abs(obj)
```

```python
my_points = [Point(1,2), Point(3,4), Point(5,6), Point(7,8)]
```

```python
def sort_list(l_source):
    l = l_source.copy()
    for _ in range(len(l)):
        for ind in range(len(l) - 1):
            if abs(l[ind + 1]) < abs(l[ind]):
                l[ind], l[ind + 1] = l[ind + 1], l[ind]
    return l
```

```python
sort_list(my_points)
```
