```python
my_str = 'abcdeaaafgabc'
def count_char(string: str) -> dict[str, int]:
    """
    Calculates frequency of each character in the input string.
    """
    char_freq = {}
    for char in string:
        if char in char_freq:
            char_freq[char] += 1
        else:
            char_freq[char] = 1
    return char_freq
```

```python
count_char(my_str)
```

```python
count_char(my_str)['a']
```

```python
subjects = ['I', 'You']
verbs = ['Play', 'Love', 'Python']
objects = ['Statistics', 'Math']
```

```python
def make_sentence(subjecs, verbs, objects):
    """
    Generate all possible sentences from given subjects, verbs, and objects.
    """
    sentences = []
    for subject in subjects:
        for verb in verbs:
            for obj in objects:
                sentences.append(f"{subject} {verb} {obj}")
    return sentences
```

```python
make_sentence(subjects, verbs, objects)
```
