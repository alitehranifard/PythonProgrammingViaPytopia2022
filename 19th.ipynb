```python
# Method 1: Using slicing (most Pythonic)
original_list = [1,2,3,4,5,6,7]
reversed_list = original_list[::-1]
print("Method 1 (slicing):", reversed_list)

# Method 2: Using reverse() method (modifies original list)
list2 = [1,2,3,4,5,6,7]
list2.reverse()
print("Method 2 (reverse method):", list2)

# Method 3: Manual reversal with indices
list3 = [1,2,3,4,5,6,7]
for i in range(len(list3) // 2):
    list3[i], list3[-(i+1)] = list3[-(i+1)], list3[i]
print("Method 3 (manual swap):", list3)

# Method 4: Using enumerate to get indices
list4 = [1,2,3,4,5,6,7]
for i, value in enumerate(list4):
    if i < len(list4) // 2:
        list4[i] = list4[-(i+1)]
        list4[-(i+1)] = value
print("Method 4 (enumerate):", list4)

```
