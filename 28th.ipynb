```python
def check_validation(*kwargs):
    for username,password in kwargs.items():
        if len(username) < 4:
            is_valid = False
        if len(password) < 6:
            is_valid = False
        if not username.isalpha():
            is_valid = False
        if  password.isdigit():     
            is_valid = False
        if password.lower() == password:
            is_valid = False
        return is_valid
```

```python
def is_valid_password(password: str) -> bool:
    """
    Checks if password is vali based on rules.
    """
    is_valid = True
    if (
        len(password)<6
        or password.isdigit()
        or password.lower() == password
    ):
        is_valid = False
    return is_valid
```

```python
def is_valid_username(username):
    is_valid_2 = True
    if len(username) < 4:
        is_valid_2 = False
    return is_valid_2
```
